```python
import pandas as pd
csv_path = 'mistral_responses.csv'
df = pd.read_csv(csv_path)
df.head()
```

```python
mistral_responses = list(df['original_response_mistral'])
final_responses = []
for resp in mistral_responses:
    ans = resp.split('\n')[0].strip().split(' ')[0]
    final_responses.append(ans)
```

```python
df['processed_mistral'] = final_responses
```

```python
df.to_csv('processed_mistral.csv')
```
